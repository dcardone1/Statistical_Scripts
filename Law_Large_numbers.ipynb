### This code is intended to show numerically the law of large numbers considering a population with some random variable X which has a standard normal distribution. The program simulates taking a sample of n individuals calculating the sample mean for X and then plot it. Iterating with a sample of increasing size ii is shown how the sample mean converges to the population mean.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Here the sample means are collected
sample_means = []
max_sample_size = 10000
step = 20

#iteration with increasing number of samples
for n in range(1, max_sample_size, step):
    sample_vector = np.random.normal(loc=0, scale=1, size=(n,))
    sample_means.append(np.mean(sample_vector))
```

```python
#variables to plot
x = np.arange(1,max_sample_size, step)
means = np.array(sample_means)

fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot()
ax.plot(x, means)
ax.hlines(y=0, xmin=0, xmax=max_sample_size, colors=['r'], linestyles=['--'])
ax.set_xlabel('Iterations')
ax.set_ylabel('Average')
ax.grid()
```
