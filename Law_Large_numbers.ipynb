### This code will generate n samples from the standar normal distribution taken as the known population distribution, it will take the mean an plot it, it will iterate with increasing values of n, as n increases it should be seen the the sample mean aproaches the population mean. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Here the sample means are collected
sample_means = []
max_sample_size = 300
step = 1

#iteration with increasing number of samples
for n in range(2, max_sample_size, step):
    sample_vector = np.random.normal(loc=0, scale=1, size=(n,))
    sample_means.append(np.mean(sample_vector))
```

```python
#variables to plot
x = np.arange(2,max_sample_size, step)
means = np.array(sample_means)

fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot()
ax.plot(x, means)
ax.hlines(y=0, xmin=0, xmax=max_sample_size, colors=['r'], linestyles=['--'])
ax.grid()
```
